```python
# From Body to Floral are the whiskey flavors
# To select only those columns which are of flavors i.e from body (index = 2) to floral (index = 13)
flavors = whisky.iloc[: , 2:14]               # Floral is included 
flavors
```

```python
len(flavors.columns)
```

```python
flavors_correlation = pd.DataFrame.corr(flavors)
flavors_correlation
```

```python
plt.figure(figsize=(10, 10))
plt.pcolor(flavors_correlation)
plt.colorbar()
plt.savefig('flavors_corr.pdf')
```

```python
whisky_correlation = pd.DataFrame.corr(flavors.transpose())
plt.figure(figsize=(10, 10))
plt.pcolor(whisky_correlation)
plt.axis('tight')
plt.colorbar()
plt.savefig("whisky_corr.pdf")
```

```python
from sklearn.cluster import SpectralCoclustering
```

```python
model = SpectralCoclustering(n_clusters=6, random_state=0, )
model.fit(whisky_correlation)
model.rows_
```

```python
np.sum(model.rows_, axis = 1)          # Shows no. of whiskies belonging to each of the 6 (0-5) clusters
```

```python
np.sum(model.rows_, axis = 0)          # No. of clusters each entry belongs to
```

```python
model.row_labels_                    # The cluster each entry belongs to
```

```python
whisky["Groups"] = pd.Series(model.row_labels_, index = whisky.index)
```

```python
whisky = whisky.iloc[np.argsort(model.row_labels_)]
```

```python
whisky.reset_index(drop = True)
```

```python
correlations = pd.DataFrame.corr(whisky.iloc[: , 2: 14].transpose())
```

```python
correlations = np.array(correlations)
```

```python
plt.figure(figsize = (14,7))
plt.subplot(121)
plt.pcolor(whisky_correlation)
plt.title("Original")
plt.axis("tight")
plt.subplot(122)
plt.pcolor(correlations)
plt.title("Rearranged")
plt.axis("tight")
plt.savefig("correlations.pdf")
```
